```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

def h(x):
    return int(f(x))


```

```python
x_min = optimize.minimize(h, 30, method='BFGS')
print(x_min)
```

```python
x_min = optimize.differential_evolution(h, [(1, 30)])
print(f(30))
print(int(-5.6122087764738))
```

```python
x = np.arange(1, 31, 0.5)
y = np.array([f(g) for g in x])
y1 = np.array([h(g) for g in x])
```
