```python
import numpy as np
import scipy as sp
import math as m
import matplotlib.pyplot as plt
```

```python
def func(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

def create_matrix(n, x):
    a = np.zeros((len(x), n+1))
    for i in range(len(x)):
        for j in range(n+1):
            a[i,j] = x[i]**j
    return a
```

```python
x1 = 1
x2 = 15
x = np.array([[x1], [x2]])
a = create_matrix(1, x)
print(a)
x_1 = np.array([[1, x1], [1, x2]])
f_1 = func(x)
w_1 = np.linalg.solve(a, f_1)
print(w_1)
y = np.array(np.dot(create_matrix(1, np.arange(0, 15, 0.1)), w_1))
```

```python
plt.plot(np.arange(0, 15, 0.1), y)
plt.show()
```

```python
x1 = 1
x2 = 8
x3 = 15
x = np.array([[x1], [x2], [x3]])
a = create_matrix(2, x)
x_2 = np.array([[1, x1, x1**2], [1, x2, x2**2], [1, x3, x3**2]])
print(func(x1))
f_2 = func(x)
w_2 = np.linalg.solve(x_2, f_2)
y = np.array(np.dot(create_matrix(2, np.arange(0, 15, 1)), w_2))
```

```python
plt.plot(np.arange(0, 15, 1), y)
plt.show()
```
